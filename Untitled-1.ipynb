```python
import numpy as np
from itertools import permutations

# ── state space ──────────────────────────────────────────────────────────────
# state s_i = permutation (r1, r2, r3) where rk = true rank of candidate k
# e.g. (1,2,3) means candidate 1 is best, 2 is second, 3 is worst

states = list(permutations([1, 2, 3]))
n_states = len(states)
state_idx = {s: i for i, s in enumerate(states)}

print("States:")
for i, s in enumerate(states):
    print(f"  s{i+1} = {s}  (cand1_rank={s[0]}, cand2_rank={s[1]}, cand3_rank={s[2]})")

```

```python
# ── P1: before seeing anyone ──────────────────────────────────────────────────
# all permutations equally likely → uniform 6x6 matrix
P1 = np.ones((n_states, n_states)) / n_states

# ── P2: after seeing candidate 1 ─────────────────────────────────────────────
# you now know relative rank of candidate 2 vs candidate 1
# group A: candidate 2 better than candidate 1  → s[1] < s[0]
# group B: candidate 1 better than candidate 2  → s[0] < s[1]
# within each group, remaining 3 permutations equally likely (prob 1/3)

P2 = np.zeros((n_states, n_states))
for i, s_from in enumerate(states):
    group_A = s_from[1] < s_from[0]   # cand2 better than cand1
    for j, s_to in enumerate(states):
        to_A = s_to[1] < s_to[0]
        if group_A == to_A:            # same group → prob 1/3
            P2[i, j] = 1/3

# ── P3: after seeing all candidates ──────────────────────────────────────────
# you know the complete ranking → identity
P3 = np.eye(n_states)

# ── print transition matrices ───────────────────────────────────────────────
print("\nTransition matrix P1 (before seeing anyone):")
print(P1)       
print("\nTransition matrix P2 (after seeing candidate 1):")
print(P2)
print("\nTransition matrix P3 (after seeing all candidates):")
print(P3)
```

```python
# ── A1: garbling such that P1 = A1 @ P2 ──────────────────────────────────────
# P2 has rank 2, so A1 is not unique
# canonical choice: uniform garbling A1 = (1/6) * ones
# verification: A1 @ P2 = (1/6)*ones @ P2 = (1/6)*ones (since col sums of P2 = 1)
A1 = np.ones((n_states, n_states)) / n_states

# ── A2: garbling such that P2 = A2 @ P3 = A2 @ I = A2 ───────────────────────
A2 = P2.copy()

# ── verification ─────────────────────────────────────────────────────────────
print("\n── Verification ─────────────────────────────────────────────────")
print(f"P1 = A1 @ P2:  max error = {np.max(np.abs(P1 - A1 @ P2)):.2e}")
print(f"P2 = A2 @ P3:  max error = {np.max(np.abs(P2 - A2 @ P3)):.2e}")
print(f"P1 = A1 @ A2 @ P3:  max error = {np.max(np.abs(P1 - A1 @ A2 @ P3)):.2e}")

# row sums
for name, M in [("A1", A1), ("A2", A2), ("P1", P1), ("P2", P2), ("P3", P3)]:
    print(f"  {name} row sums all 1: {np.allclose(M.sum(axis=1), 1)}")

# ── print matrices ────────────────────────────────────────────────────────────
def print_matrix(name, M):
    print(f"\n{name}  ({M.shape[0]}x{M.shape[1]}):")
    # format as fractions for readability
    from fractions import Fraction
    for row in M:
        fracs = [str(Fraction(v).limit_denominator(100)) for v in row]
        print("  [" + "  ".join(f"{f:>4}" for f in fracs) + "]")

# print_matrix("P1", P1)
# print_matrix("P2", P2)
# print_matrix("P3", P3)
print_matrix("A1", A1)
print_matrix("A2", A2)

```

```python
# ── rank ──────────────────────────────────────────────────────────────────────
print("\n── Ranks ────────────────────────────────────────────────────────")
for name, M in [("P1", P1), ("P2", P2), ("P3", P3), ("A1", A1), ("A2", A2)]:
    print(f"  rank({name}) = {np.linalg.matrix_rank(M)}")

# ── group structure ───────────────────────────────────────────────────────────
print("\n── Group structure at stage 2 ───────────────────────────────────")
groupA = [i for i, s in enumerate(states) if s[1] < s[0]]
groupB = [i for i, s in enumerate(states) if s[0] < s[1]]
print(f"  Group A (cand2 better): {[f's{i+1}={states[i]}' for i in groupA]}")
print(f"  Group B (cand1 better): {[f's{i+1}={states[i]}' for i in groupB]}")

```

```python
# ── cascade examples ──────────────────────────────────────────────────────────
print("\n\n── Example 1: Delta_A1 that does NOT cascade to A2 ─────────────")
# perturb within Group B: move mass between s1=(1,2,3) and s2=(1,3,2)
# these two states have IDENTICAL rows in P2 → P2 cannot distinguish them
# so any perturbation swapping mass between s1 and s2 is invisible to A2

# A1' assigns deterministic mass within each group
# row i goes to exactly one state within its group
A1_prime_case1 = np.zeros((n_states, n_states))
for i, s in enumerate(states):
    group_A = s[1] < s[0]
    if group_A:
        # send all mass to s3 = (2,1,3) within group A
        A1_prime_case1[i, state_idx[(2,1,3)]] = 1/3
        A1_prime_case1[i, state_idx[(3,1,2)]] = 1/3  
        A1_prime_case1[i, state_idx[(3,2,1)]] = 1/3
    else:
        # send all mass to s1 = (1,2,3) within group B
        A1_prime_case1[i, state_idx[(1,2,3)]] = 1/3
        A1_prime_case1[i, state_idx[(1,3,2)]] = 1/3
        A1_prime_case1[i, state_idx[(2,3,1)]] = 1/3
print_matrix("A1", A1)
print_matrix("A1'", A1_prime_case1)
print(f"  rank(A1') = {np.linalg.matrix_rank(A1_prime_case1)}")
print("know the relative rank of first canddidate and second before starting the interview")
print(A1_prime_case1 @ P2)
```

```python
# original A1: uniform 1/6 everywhere (rank 1)
# new A1': still rank 1 (all rows identical)
# but now biased toward Group B

# Group B states: s1=(1,2,3), s2=(1,3,2), s4=(2,3,1)  → indices 0,1,3
# Group A states: s3=(2,1,3), s5=(3,1,2), s6=(3,2,1)  → indices 2,4,5

# assign weight p to Group B (p > 1/2 means more confident cand1 better)
# within each group still uniform

p = 0.7   # 70% sure candidate 1 is better than candidate 2

A1_prime = np.zeros((n_states, n_states))
groupB_idx = [state_idx[(1,2,3)], state_idx[(1,3,2)], state_idx[(2,3,1)]]
groupA_idx = [state_idx[(2,1,3)], state_idx[(3,1,2)], state_idx[(3,2,1)]]

for i in range(n_states):
    for j in groupB_idx:
        A1_prime[i, j] = p / 3        # weight p split equally within Group B
    for j in groupA_idx:
        A1_prime[i, j] = (1-p) / 3   # weight (1-p) split equally within Group A
print_matrix("A1'", A1_prime)
print(f"  rank(A1') = {np.linalg.matrix_rank(A1_prime)}")
print(A1_prime @ P2)
print("\nA1' row 0 (all rows identical):")
for j, s in enumerate(states):
    grp = "B" if j in groupB_idx else "A"
    print(f"  s{j+1}={states[j]} Group {grp}: weight = {A1_prime[0,j]:.4f}")
```

```python
"""
Secretary Problem (n=3): Dual LP Solution
==========================================
Solves M-LP (dual of V-LP) for optimal stopping:

  max  sum_t lambda_t^T Q_t C_t
  s.t. p_0          = lambda_1 + mu_1          [t=1]
       P_{t-1}^T mu_{t-1} = lambda_t + mu_t    [t=2]
       P_2^T mu_2   = lambda_3                  [t=3]
       lambda_t, mu_t >= 0

Then recovers optimal strategy: pi_t(g) = lambda_t(g) / (lambda_t(g) + mu_t(g))
"""

import numpy as np
from scipy.optimize import linprog
import itertools

# ── States ────────────────────────────────────────────────────────────────────
perms     = list(itertools.permutations([1,2,3]))
S         = len(perms)                          # 6
state_idx = {p: i for i, p in enumerate(perms)}

# ── Information partitions ────────────────────────────────────────────────────
def relative_rank(perm, t):
    """Rank of candidate t among first t seen (1 = best so far)"""
    return sum(1 for r in perm[:t] if r <= perm[t-1])

def build_partition(t):
    groups = {}
    for perm, idx in state_idx.items():
        groups.setdefault(relative_rank(perm, t), []).append(idx)
    keys = sorted(groups)
    return [groups[k] for k in keys], keys

partitions, part_keys = {}, {}
for t in [1,2,3]:
    partitions[t], part_keys[t] = build_partition(t)

# ── Matrices ──────────────────────────────────────────────────────────────────
def build_Q(t):                         # R^{|n_t| x S}: averaging within groups
    Q = np.zeros((len(partitions[t]), S))
    for g, grp in enumerate(partitions[t]):
        for s in grp: Q[g,s] = 1/len(grp)
    return Q

def build_P(t):                         # R^{|n_t| x |n_{t+1}|}: group transitions
    P = np.zeros((len(partitions[t]), len(partitions[t+1])))
    for i,gi in enumerate(partitions[t]):
        for j,gj in enumerate(partitions[t+1]):
            inter = set(gi)&set(gj)
            if inter: P[i,j] = len(inter)/len(gi)
    return P

Q = {t: build_Q(t) for t in [1,2,3]}
P = {t: build_P(t) for t in [1,2]}

# ── Rewards ───────────────────────────────────────────────────────────────────
# C_t(s) = 1 if candidate t has true rank 1, else 0
C  = {t: np.array([1.0 if perms[i][t-1]==1 else 0.0 for i in range(S)])
      for t in [1,2,3]}
QC = {t: Q[t] @ C[t] for t in [1,2,3]}   # expected reward if stop at t, per group

# ── Initial distribution ──────────────────────────────────────────────────────
# p0(g) = P(state in group g) = |g|/S  (uniform over states)
p0 = np.array([len(grp)/S for grp in partitions[1]])  # sums to 1

# ── Variable layout ───────────────────────────────────────────────────────────
n1,n2,n3 = len(partitions[1]), len(partitions[2]), len(partitions[3])
# [lambda_1|mu_1|lambda_2|mu_2|lambda_3]
sl1 = slice(0,     n1)
sm1 = slice(n1,    2*n1)
sl2 = slice(2*n1,  2*n1+n2)
sm2 = slice(2*n1+n2, 2*n1+2*n2)
sl3 = slice(2*n1+2*n2, 2*n1+2*n2+n3)
nv  = 2*n1+2*n2+n3

# ── Objective: max sum_t lambda_t^T QC_t ─────────────────────────────────────
c_obj = np.zeros(nv)
c_obj[sl1] = -QC[1];  c_obj[sl2] = -QC[2];  c_obj[sl3] = -QC[3]

# ── Equality constraints ──────────────────────────────────────────────────────
Aeq = np.zeros((n1+n2+n3, nv));  beq = np.zeros(n1+n2+n3)

# t=1: lambda_1 + mu_1 = p0
Aeq[:n1, sl1] = np.eye(n1);  Aeq[:n1, sm1] = np.eye(n1);  beq[:n1] = p0

# t=2: P_1^T mu_1 - lambda_2 - mu_2 = 0
Aeq[n1:n1+n2, sm1] = P[1].T
Aeq[n1:n1+n2, sl2] = -np.eye(n2)
Aeq[n1:n1+n2, sm2] = -np.eye(n2)

# t=3: P_2^T mu_2 - lambda_3 = 0
Aeq[n1+n2:, sm2] = P[2].T;  Aeq[n1+n2:, sl3] = -np.eye(n3)

# ── Solve ─────────────────────────────────────────────────────────────────────
res = linprog(c_obj, A_eq=Aeq, b_eq=beq, bounds=[(0,None)]*nv, method='highs')
x   = res.x

lam = {1: x[sl1], 2: x[sl2], 3: x[sl3]}
mu  = {1: x[sm1], 2: x[sm2]}

# ── Print results ─────────────────────────────────────────────────────────────
print("╔══════════════════════════════════════════════════════╗")
print("║   SECRETARY PROBLEM n=3 — DUAL LP SOLUTION          ║")
print("╚══════════════════════════════════════════════════════╝")

print(f"\nOptimal value = {-res.fun:.6f}   (known: 1/2 = {0.5:.6f})")

print("\n── Dual variables ──────────────────────────────────────")
for t in [1,2,3]:
    keys = part_keys[t]
    l    = lam[t]
    m    = mu[t] if t<3 else np.zeros(n3)
    print(f"\nt={t}  (rel_rank groups: {keys})")
    print(f"  lambda_{t} = {np.round(l,4)}")
    if t < 3:
        print(f"  mu_{t}     = {np.round(m,4)}")
    print(f"  y_{t}      = {np.round(l+m,4)}   (total mass at each group)")

print("\n── Flow conservation ───────────────────────────────────")
print(f"  t=1: |lam1+mu1 - p0|    = {np.abs(lam[1]+mu[1]-p0).max():.2e}")
print(f"  t=2: |P1^T mu1 - lam2-mu2| = {np.abs(P[1].T@mu[1]-lam[2]-mu[2]).max():.2e}")
print(f"  t=3: |P2^T mu2 - lam3|  = {np.abs(P[2].T@mu[2]-lam[3]).max():.2e}")

print("\n── Optimal stopping strategy ───────────────────────────")
print("  pi_t(g) = lambda_t(g) / y_t(g)\n")
for t in [1,2,3]:
    print(f"  t={t}:")
    l = lam[t]
    m = mu[t] if t<3 else np.zeros(n3)
    y = l+m
    for key, li, mi, yi in zip(part_keys[t], l, m, y):
        pi  = li/yi if yi>1e-10 else (1.0 if t==3 else 0.0)
        act = "STOP" if pi>0.5 else "CONTINUE"
        print(f"    rel_rank={key}: pi={pi:.4f} → {act}")

# print("\n── Interpretation ──────────────────────────────────────")
# print("  t=1: always CONTINUE (never hire first candidate)")
# print("  t=2: STOP if candidate 2 is best so far (rel_rank=1)")
# print("       CONTINUE otherwise")
# print("  t=3: always STOP (forced)")
# print("\n  This is exactly the classical n=3 secretary rule:")
# print("  'Skip the first candidate, then hire the first")
# print("   who beats all previously seen.'")

# print("\n── Sensitivity at each stage ───────────────────────────")
# print("  delta_val ≈ lambda_t^T delta_Q_t C_t + mu_t^T delta_P_t V*_{t+1}")
# print("  Refinement is harmless iff:")
# print("  (1) delta_Q_t^T lambda_t = 0  (stop nodes: reward already measurable)")
# print("  (2) delta_P_t^T mu_t = 0      (continue nodes: transition already measurable)")
for t in [1,2]:
    l_active = [(key,li) for key,li in zip(part_keys[t],lam[t]) if li>1e-10]
    m_active = [(key,mi) for key,mi in zip(part_keys[t],mu[t])  if mi>1e-10]
    print(f"\n  t={t}: stopping nodes (lambda>0): {l_active}")
    print(f"        continuation nodes (mu>0):   {m_active}")
```
